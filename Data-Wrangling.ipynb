```python
#Lets clean up and compare the attack column when we display data that excludes the outliers
attack = df['Attack']
q1 = attack.quantile(.25)
q3 = attack.quantile(.75) 
iqr = q3 - q1
# All values less than pmin and greater than pmax are outliers
pmin = q1 - 1.5 * iqr
pmax = q3 + 1.5 * iqr
non_outliers = attack.where(attack.between(pmin,pmax))
```

```python
plt.boxplot(attack)
```

```python
compare = pd.DataFrame({'before': attack, 'after': non_outliers})
compare.plot.box()
compare.describe()
```
